```python
Q1=np.percentile(df['runs_last_5'],25,interpolation='midpoint')
Q2=np.percentile(df['runs_last_5'],50,interpolation='midpoint')
Q3=np.percentile(df['runs_last_5'],75,interpolation='midpoint')
print(Q1)
print(Q2)
print(Q3)
```

```python
IQR=Q3-Q1
print(IQR)
```

```python
upper_limit=Q3+1.5*IQR
lower_limit=Q1-1.5*IQR
print(upper_limit)
print(lower_limit)
```

```python
outlier=[]
for x in df['runs_last_5']:
  if((x>upper_limit) or (x<lower_limit)):
    outlier.append(x)
outlier
```

```python
Q1=np.percentile(df['wickets_last_5'],25,interpolation='midpoint')
Q2=np.percentile(df['wickets_last_5'],50,interpolation='midpoint')
Q3=np.percentile(df['wickets_last_5'],75,interpolation='midpoint')
print(Q1)
print(Q2)
print(Q3)
```

```python
IQR=Q3-Q1
IQR
```

```python
upper_limit=Q3+1.5*IQR
lower_limit=Q1-1.5*IQR
print(upper_limit)
print(lower_limit)
```

```python
outlier=[]
for i in df['wickets_last_5']:
    if((i>upper_limit)or(i<lower_limit)):
        outlier.append(i)
outlier
```

```python
a1=((df["wickets_last_5"]>upper_limit)|(df["wickets_last_5"]<lower_limit))
a1
```

# KNN MODEL

```python
x=output.drop('total',axis=1)
y=output['total']
print(y)
print(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsRegressor  
from sklearn.metrics import mean_squared_error,r2_score
metric_k= []
neighbors = np.arange(2, 15)

for k in neighbors:
    knn=KNeighborsRegressor(n_neighbors=k)
    model2=knn.fit(x_train,y_train)
    y_pred_knn=model2.predict(x_test)
    accuracy=r2_score(y_test,y_pred_knn)
    metric_k.append(accuracy)
```

```python
metric_k
```

```python
plt.plot(neighbors,metric_k,'o--',color='r')
plt.xlabel('K Value')
plt.ylabel('Accuracy')
plt.grid()
```

```python
knn=KNeighborsRegressor(n_neighbors=2)
model_knn=knn.fit(x_train,y_train)
y_pred_knn=model_knn.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
print(mean_squared_error(y_test,y_pred_knn))
```

```python
print(r2_score(y_test,y_pred_knn))
```

# Performance Analysis

```python
lg_score  =51 # Linear Regression
knn_score =95 # K nearest
svm_scodre=53 # support vector
dt_score  =92 # decision tree
rf_score  =97 # random forest
xgb_score =93 # xg boost
```

```python
algorithms = ['Lin Reg','KNN' ,'SVM','DT','Random Forest','XGB']
scores = [lg_score,knn_score,svm_scodre,dt_score,rf_score,xgb_score]

# Plotting the bar graph 
plt.figure(figsize=(10, 5))
plt.bar(algorithms, scores, color=['yellow', 'orange', 'green', 'pink', 'red','cyan'])
plt.xlabel("Algorithms")
plt.ylabel("R2score")
plt.title("Performance Analysis of Different Regression Algorithms")
plt.ylim(0, max(scores) + 1) 
plt.grid(axis='y',alpha=0.5)
plt.show()
```
